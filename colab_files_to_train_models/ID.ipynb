```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.svm import SVC
```

Data collection & analysis

```python
id = pd.read_csv('/content/processed_vineland_data.csv')
id.head()
```

```python
id.shape
```

```python
id.isnull().sum()
```

```python
# Fill null values in 'Intellectual_Disability_Level' with the most frequent value
most_frequent_level = id['Intellectual_Disability_Level'].mode()[0]
id['Intellectual_Disability_Level'].fillna(most_frequent_level, inplace=True)
```

```python
id.info()
```

```python
id.describe()
```

```python
id['Intellectual_Disability_Level'].value_counts()
```

```python
# Calculate the mean for numeric columns only
id.groupby('Intellectual_Disability_Level').mean(numeric_only=True)
```

```python
id.columns
```

```python
# Separate features (X) and target (y)
X = id[[ 'Communication_Score', 'Daily_Living_Score', 'Socialization_Score','Overall_Adaptive_Score']]
Y = id['Intellectual_Disability_Level']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Create and train the model (Support Vector Machine with regularization)
model_svm = SVC(kernel='linear', C=0.1, random_state=42)  # Adjust C for regularization strength
model_svm.fit(X_train, Y_train)
```

```python
# Make predictions on the test set
y_pred_svm = model_svm.predict(X_test)
```

```python
# Evaluate the model on the training set
y_pred_train_svm = model_svm.predict(X_train)
accuracy_train_svm = accuracy_score(Y_train, y_pred_train_svm)
print("Training Accuracy (SVM with Regularization):", accuracy_train_svm)
```

```python
# Evaluate the model on the testing set (already done)
accuracy_svm = accuracy_score(Y_test, y_pred_svm)
print("Testing Accuracy (SVM with Regularization):", accuracy_svm)
```

```python
input_data = [ 'Communication_Score', 'Daily_Living_Score', 'Socialization_Score','Overall_Adaptive_Score']
input_data = [80, 70, 65, 75]  # Example input values

# Convert input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Make prediction using the trained model
prediction_intellectual = model_svm.predict(input_data_reshaped)

# Print the prediction
print("Prediction for Intellectual Disability Level:", prediction_intellectual[0])

# You can add more code to interpret the prediction and display it user-friendly.
# Example:
if prediction_intellectual[0] == 'Mild':
  print('The person has mild intellectual disability')
elif prediction_intellectual[0] == 'Moderate':
  print('The person has moderate intellectual disability')
elif prediction_intellectual[0] == 'Severe':
  print('The person has severe intellectual disability')
else:
  print('The person does not have intellectual disability')
```
